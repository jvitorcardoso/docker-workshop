```sql
SELECT
  vendorid AS vendorId,
  DATE(tpep_pickup_datetime) :: STRING || ' ' || LPAD(HOUR(tpep_pickup_datetime), 2, '0')  || ':' || LPAD(MINUTE(tpep_pickup_datetime), 2, '0') || ':' || LPAD(SECOND(tpep_pickup_datetime), 2, '0') AS pickupHour,
  DATE(tpep_dropoff_datetime) :: STRING || ' ' || LPAD(HOUR(tpep_dropoff_datetime), 2, '0')  || ':' || LPAD(MINUTE(tpep_dropoff_datetime), 2, '0') || ':' || LPAD(SECOND(tpep_dropoff_datetime), 2, '0') AS pickupHour,
  TIMESTAMPDIFF(MINUTE, tpep_pickup_datetime, tpep_dropoff_datetime) AS rideDuration,
  passenger_count AS passengerCount,
  trip_distance AS tripDistance,
  ratecodeid AS rateCodeId,
  store_and_fwd_flag AS storeAndFwdFlag,
  pulocationid AS pickupLocationId,
  dolocationid AS dropoffLocationId,
  payment_type AS paymentType,
  fare_amount AS fareAmount,
  extra AS extra,
  mta_tax AS mtaTax,
  tip_amount AS tipAmount,
  tolls_amount AS tollsAmount,
  total_amount AS totalAmount,
  congestion_surcharge AS congestionSurcharge
FROM `docker-workshop`.public.yellow_taxi_data_2021_1
WHERE
  TIMESTAMPDIFF(MINUTE, tpep_pickup_datetime, tpep_dropoff_datetime) >= 30
  AND trip_distance > 5
  AND tolls_amount > 0
  AND vendorid IS NOT NULL;
```
